# data validation

```python
# validate songs_df
```

```python
def validate_songs_df(songs_df):
    # drop rows with missing critical values
    songs_df.dropna(subset=['popularity', 'track_id', 'artists', 'track_name', 'duration_ms'], inplace=True)
    # fill nan values with 0
    songs_df.fillna(0, inplace=True)
    # drop duplicate rows
    songs_df.drop_duplicates(inplace=True)

    # Validate the range for numerical columns
    songs_df = songs_df[(songs_df['duration_ms'] > 0)]  # duration_ms must be positive
    songs_df = songs_df[(songs_df['danceability'].between(0, 1))]  # between 0 and 1
    songs_df = songs_df[(songs_df['energy'].between(0, 1))]
    songs_df = songs_df[(songs_df['loudness'].between(-80, 5))]  # loudness should be in a reasonable range
    songs_df = songs_df[(songs_df['mode'].isin([1, 0]))] # mode or 0 or 1
    songs_df = songs_df[(songs_df['speechiness'].between(0, 1))] #
    songs_df = songs_df[(songs_df['acousticness'].between(0, 1))] #
    songs_df = songs_df[(songs_df['instrumentalness'].between(0, 1))] #
    songs_df = songs_df[(songs_df['liveness'].between(0, 1))] #
    songs_df = songs_df[(songs_df['valence'].between(0, 1))] #
    songs_df = songs_df[(songs_df['tempo'] > 0)]  # tempo must be positive

    return songs_df
```

```python
# validate users_df
```

```python
def validate_users_df(users_df):
    # drop rows with missing critical values
    users_df.dropna(subset=['user_id', 'user_name'], inplace=True)
    # fillna with 0
    users_df.fillna(0, inplace=True)
    # created_at to date
    users_df['created_at'] = pd.to_datetime(users_df['created_at'], errors='coerce')
    users_df.dropna(subset=['created_at'], inplace=True)
    # drop duplicates
    users_df.drop_duplicates(subset=['user_id'], inplace=True)

    return users_df
```

```python
# streams validation
```

```python
def validate_stream_df(streams_df, songs_df, users_df):
    # Ensure that track_id in stream_df exists in songs_df
    streams_df = streams_df[streams_df['track_id'].isin(songs_df['track_id'])]
    # Ensure that user_id in stream_df exists in users_df
    streams_df = streams_df[streams_df['user_id'].isin(users_df['user_id'])]
    # convert date
    streams_df['listen_time'] = pd.to_datetime(streams_df['listen_time'], errors='coerce') 
    streams_df.dropna(subset=['listen_time'], inplace=True)

    return streams_df
```

```python
users_df = validate_users_df(users_df)
```

```python
songs_df = validate_songs_df(songs_df)
```

```python
streams_total_df = validate_stream_df(streams_total_df, songs_df, users_df)
```

## calculate metrics and listening patterns

```python
# 1) Top Genres by Streams and Age group
```

```python
def calculate_genres_age_distribution(users_df, songs_df, streams_total_df):
    age_bins = [0, 18, 25, 35, 45, 100]
    age_labels = ['Under 18', '18-25', '26-35', '36-45', '46+']
    users_df['age_group'] = pd.cut(users_df['user_age'], bins=age_bins, labels=age_labels, right=False)

    merged_df = pd.merge(streams_total_df, users_df, on='user_id', how='inner')
    merged_df = pd.merge(merged_df, songs_df, on='track_id', how='inner')
    count_streams = merged_df.groupby(['age_group', 'track_genre'], observed=True).size().reset_index(name='stream_count')

    return count_streams
```

```python
count_streams = calculate_genres_age_distribution(users_df, songs_df, streams_total_df)
```

```python
count_streams.sort_values(by=["stream_count"], ascending=False).head(10)
```

```python
# 2) Peak Streaming Hours by country
```

```python
def calculate_country_hours_distribution(songs_df, users_df, streams_total_df):
    merged_df = pd.merge(streams_total_df, users_df, on='user_id', how='inner')
    hours_country_df = merged_df.groupby(by=[merged_df.listen_time.dt.hour, 'user_country']).size().reset_index(name='stream_count')
    return hours_country_df
```

```python
hours_country_df = calculate_country_hours_distribution(songs_df, users_df, streams_total_df)
```

```python
hours_country_df.sort_values(by=["listen_time", "stream_count"], ascending=[True, False])
```

```python
from airflow import DAG
from airflow.providers.amazon.aws.hooks.s3 import S3Hook
from airflow.operators.python import PythonOperator
from airflow.utils.dates import days_ago
```

```python
BUCKET_NAME = "spsongsbucket"
PREFIX = "data/"
```

```python
REDSHIFT_TABLE = "user_data"
```

```python
def read_csvs_from_s3(**kwargs):
    """ Reads all CSV files from S3 and saves their content in XCom """
    s3_hook = S3Hook(aws_conn_id="aws_default")
    files = s3_hook.list_keys(bucket_name=BUCKET_NAME, prefix=PREFIX)

    if not files:
        print("No files found in S3.")
        return []

    csv_data = []
    
    for file_key in files:
        if file_key.endswith("users.csv"):  # Ensure we're only reading CSVs
            file_content = s3_hook.read_key(bucket_name=BUCKET_NAME, key=file_key)
            csv_data.append({"file_name": file_key, "content": file_content})
    
    print(f"Read {len(csv_data)} CSV files from S3.")
    
    return csv_data  # Save to XCom
```

```python
csv_data = read_csvs_from_s3()
```

```python
from airflow.providers.postgres.hooks.postgres import PostgresHook
import pandas as pd
from io import StringIO
```

```python
def load_csv_to_redshift(**kwargs):
    """ Gets CSV data from XCom and loads it into Redshift """
    ti = kwargs["ti"]
    csv_data = ti.xcom_pull(task_ids="read_s3_task")  # Fetch from XCom
    
    if not csv_data:
        print("No data found in XCom.")
        return
    
    redshift_hook = PostgresHook(postgres_conn_id="redshift_connection")
    conn = redshift_hook.get_conn()
    cursor = conn.cursor()
    
    for file in csv_data:
        print(f"Processing {file['file_name']}")
        
        # Load CSV into Pandas DataFrame
        df = pd.read_csv(StringIO(file["content"]))
        
        # Insert into Redshift
        for _, row in df.iterrows():
            insert_query = f"""
            INSERT INTO {REDSHIFT_TABLE} (user_id, user_name, user_age, user_country, created_at)
            VALUES (%s, %s, %s, %s, %s)
            """
            cursor.execute(insert_query, (
                row["user_id"],
                row["user_name"],
                row["user_age"],
                row["user_country"],
                row["created_at"]
            ))

    conn.commit()
    cursor.close()
    conn.close()
    print("Data successfully inserted into Redshift.")
```
